```python
working_dir = '/Volumes/bog_archive/DataArchives/Regions/NortheastPacific/CruiseData/aMBTS/Raw_stoqs_lrauv_data/'
print(working_dir)
```

```python
files_to_fix = []
for folder, sub_folders, files in os.walk(os.getcwd()):
    for file in files:
#         print(file)
        if file.split('_')[-1] == 'transposed.csv':
            pass
        elif '~$' in file:
            pass
        elif '.DS_Store' in file:
            pass
        else:
            full_path = folder + '/' + file
            files_to_fix.append(full_path)
        

print(files_to_fix)
```

```python
files_to_fix
```

```python

for file in files_to_fix:
    
    print(file)
    #read in df
    
    file_name = file.split('/')[-1].split('.csv')[0]
    end_file_name = file_name + '_transposed.csv'
    df = pd.read_csv(file, index_col=False)
    print('\nNow working on file: ' + file_name)
    print(end_file_name)
    print('end file location: ' + working_dir + 'stoqs_lrauv_data_transposed/'+end_file_name)
    
    # set up the dataframe
    params = df['parameter__name'].drop_duplicates()
#     params = df.iloc[:,3].drop_duplicates()
    cols = ['activity__name','activity__startdate','activity__enddate','activity__start_date','activity__start_time','activity__end_date','activity__end_time','activity__mindepth','activity__maxdepth','activity__meandepth','activity__mappoint','activity__mappoint.x','activity__mappoint.y']
    all_cols = pd.DataFrame({'bottle':[]})
    for col in cols:
        all_cols[col] = []
    for param in params:
        all_cols[param] = []
    bottles = df['activity__name'].drop_duplicates()
    
    for bottle in bottles:
        for col in cols:
            if col in ['activity__start_date','activity__start_time','activity__end_date','activity__end_time','activity__meandepth']:
                pass
            else:
                all_cols[col] = df[col]
    new_df = all_cols.drop_duplicates()
    new_df = new_df.copy()
    
    
    test_df = new_df.copy()
    for bottle in bottles: 
        sip_bot = df.loc[(df['activity__name']==bottle)]
        for param in params:
            sip_bot_measurement = sip_bot.loc[(sip_bot['parameter__name'] == param)]
            param_val = sip_bot_measurement['mean'].mean()
#             print(param_val)
            test_df.loc[(test_df['activity__name'] == bottle),param] = str(param_val)
    test_df['bottle'] = test_df.activity__name.str.extract(r'(.*?)[ ](\d*)')[1]
    test_df['activity__meandepth'] = (test_df['activity__mindepth'] + test_df['activity__maxdepth'])/2
    test_df['activity__start_date'] = test_df.activity__startdate.str.extract(r'(\d{4}-\d{2}-\d{2})(\w)(\d{2}:\d{2}:\d{2})')[0]
    test_df['activity__start_time'] = test_df.activity__startdate.str.extract(r'(\d{4}-\d{2}-\d{2})(\w)(\d{2}:\d{2}:\d{2})')[2]
    test_df['activity__end_date'] = test_df.activity__enddate.str.extract(r'(\d{4}-\d{2}-\d{2})(\w)(\d{2}:\d{2}:\d{2})')[0]
    test_df['activity__end_time'] = test_df.activity__enddate.str.extract(r'(\d{4}-\d{2}-\d{2})(\w)(\d{2}:\d{2}:\d{2})')[2]

    test_df
    
#     # remove the index then save file
    test_df.set_index('bottle', inplace=True)

    test_df.to_csv(working_dir + 'stoqs_lrauv_data_transposed/'+end_file_name)
```
